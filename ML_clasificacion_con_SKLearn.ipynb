#**Machine Learning: Clasificación con SKLearn**

# Introducción a la clasificación

```python
# features 1 = si y 0 = no, 1 para perro 0 para gato
#tiene pelo largo?
#tiene las uñas afiladas?
#hace miau?

perro1=[0,1,1]
perro2=[1,0,1]
perro3=[1,1,1]

gato1=[0,1,0]
gato2=[0,1,1]
gato3=[1,1,0]

x_train = [perro1,perro2,perro3,gato1,gato2,gato3] #datos
y_train = [1,1,1,0,0,0] #clases
```

El 'x' se refiere al atributo y el 'y' a nuestro target o clasificación

```python
from sklearn.svm import LinearSVC #Estimador lineal

model = LinearSVC()
model.fit(x_train,y_train) #Introduzco datos y clases, y lo entreno
```

```python
animal_misterioso = [1,1,1]
model.predict([animal_misterioso]) #Haz una prediccion
```

El modelo me dice que es un perro

```python
misterio1 = [1,1,1]
misterio2 = [1,1,0]
misterio3 = [0,1,1]

x_test = [misterio1, misterio2, misterio3] #pruebas
y_test = [0,1,1] #clase_pruebas
```

```python
previsiones = model.predict(x_test)
```

El modelo predice que el que gato es un perro, el que es un perro puso que es un gato y solo acertó la ultima prueba

```python
correctos = (previsiones == y_test).sum()
total = len(x_test)
tasa_de_acierto = correctos / total
print(f'La tasa de acierto fue de: {round(tasa_de_acierto*100,2)}%')
```

```python
from sklearn.metrics import accuracy_score #La biblioteca provee un metodo para calcular la exactitud

tasa_de_acierto = accuracy_score(y_test, previsiones) #Recibe el valor real y el valor predecido
print(f'La tasa de acierto fue de: {round(tasa_de_acierto*100,2)}%')
```
