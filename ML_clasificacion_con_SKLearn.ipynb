# Introducción a la clasificación

```python
# features 1 = si y 0 = no, 1 para perro 0 para gato
#tiene pelo largo?
#tiene las uñas afiladas?
#hace miau?

perro1=[0,1,1]
perro2=[1,0,1]
perro3=[1,1,1]

gato1=[0,1,0]
gato2=[0,1,1]
gato3=[1,1,0]

x_train = [perro1,perro2,perro3,gato1,gato2,gato3] #datos
y_train = [1,1,1,0,0,0] #clases
```

El 'x' se refiere al atributo y el 'y' a nuestro target o clasificación

```python
from sklearn.svm import LinearSVC #Estimador lineal

model = LinearSVC()
model.fit(x_train,y_train) #Introduzco datos y clases, y lo entreno
```

```python
animal_misterioso = [1,1,1]
model.predict([animal_misterioso]) #Haz una prediccion
```

El modelo me dice que es un perro

```python
misterio1 = [1,1,1]
misterio2 = [1,1,0]
misterio3 = [0,1,1]

x_test = [misterio1, misterio2, misterio3] #pruebas
y_test = [0,1,1] #clase_pruebas
```

```python
previsiones = model.predict(x_test)
```

El modelo predice que el que gato es un perro, el que es un perro puso que es un gato y solo acertó la ultima prueba

```python
correctos = (previsiones == y_test).sum()
total = len(x_test)
tasa_de_acierto = correctos / total
print(f'La tasa de acierto fue de: {round(tasa_de_acierto*100,2)}%')
```

```python
from sklearn.metrics import accuracy_score #La biblioteca provee un metodo para calcular la exactitud

tasa_de_acierto = accuracy_score(y_test, previsiones) #Recibe el valor real y el valor predecido
print(f'La tasa de acierto fue de: {round(tasa_de_acierto*100,2)}%')
```

# Pruebas replicables, estratificación, lectura de datos de internet

```python
import pandas as pd
url = 'https://gist.githubusercontent.com/ahcamachod/38673f75b54ec62ffc290eff8e7c716e/raw/6eaa07e199d9f668bf94a034cb84dac58c82fa4f/tracking.csv'
datos = pd.read_csv(url)
datos.sample(5)
```

```python
mapa = {
    'home': 'principal',
    'how_it_works': 'como_funciona',
    'contact': 'contacto',
    'bought': 'compro'
}

datos = datos.rename(columns = mapa)
datos.sample(3)
```

```python
x = datos[['principal','como_funciona','contacto']]
x.head()
```

```python
y = datos.compro
y.head()
```

```python
datos.shape #forma del dataset
```

```python
x_train = x[:75]
y_train = y[:75]
x_test = x[75:]
y_test = y[75:]
```

```python
y_train.value_counts()
```

```python
27 / 75 #Entreno con una proporción de muestras de clases de que si compró del 36% del todo conjunto de datos para entrenamiento
```

```python
y_test.value_counts()
```

```python
6 / 24 #Al momento de probar tomó el 25% solamente, lo que se ve es que hay una diferencia de proporciones
```

```python
print(f'Entrenaremos con {len(x_train)} elementos y probaremos con {len(x_test)} elementos')
```

```python
from sklearn.svm import LinearSVC 
from sklearn.metrics import accuracy_score

model = LinearSVC()
model.fit(x_train,y_train)
previsiones = model.predict(x_test)

tasa_de_acierto = accuracy_score(y_test, previsiones)
print(f'La tasa de acierto fue de: {round(tasa_de_acierto*100,2)}%')
```

No sabemos si es bueno o malo esta tasa porque no hay con que compararlo

```python
#Otra forma, se tiene que establecer un estado de aleatoriedad para que se pueda replicar el resultado siempre que se ejecute
from sklearn.model_selection import train_test_split

SEED = 42 #Ejemplo

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.25, random_state = SEED)

model = LinearSVC()
model.fit(x_train,y_train)
previsiones = model.predict(x_test)

tasa_de_acierto = accuracy_score(y_test, previsiones)
print(f'La tasa de acierto fue de: {round(tasa_de_acierto*100,2)}%')
```

```python
#Teniendo en cuenta la proporcion al momento de entrenar y al momento de probar
from sklearn.model_selection import train_test_split

SEED = 42 #Ejemplo

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.25, random_state = SEED, stratify = y)

model = LinearSVC()
model.fit(x_train,y_train)
previsiones = model.predict(x_test)

tasa_de_acierto = accuracy_score(y_test, previsiones)
print(f'La tasa de acierto fue de: {round(tasa_de_acierto*100,2)}%')
```
